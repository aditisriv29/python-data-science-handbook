### Array Indexing: Accessing Single Elements

```python
x1 = [4, 5, 6] # one dimentional array

x1[2] # last number of the array 
x1[1] # fetching number 5
```

```python
import numpy as np
x2 = np.array([
    [3, 5, 6],
    [2, 6, 8]
])
x2[1, -1] = 9
x2
```

#### **Array Slicing: Accessing Subarrays**

```python
# One dimensional Array
x = np.array([2, 3, 5, 7, 8])

x[:5] # first 5 Elements
x[5:] # all elements after the 5th element
x[5:7] # middle subarray
```

```python
# Multidimensional subarrays

x2 = np.array([

    [12, 5, 2, 4],
    [7, 6, 8, 8],
    [1, 6, 7, 7]
])
x2[0:1]
x2[:2, :3]  # two rows, three columns
x2[:3, ::2]  # all rows, every other column
arr = x2[::-1, ::-1] # subarray dimensions wil be reversed
```

```python
print(x2[:, 0])  # first column of x2
print(x2[0, :])  # first row of x2
print(x2[0])  # equivalent to x2[0, :]
```

### **Reshaping of Arrays**

```python
import numpy as np

# reshape = change shape, not data
np.arange(1, 10).reshape(3, 3)

# 1D â†’ 2D
x = np.array([1, 2, 3])
x.reshape(1, 3)      # row
x.reshape(3, 1)      # column
x[np.newaxis, :]     # row (alt)
x[:, np.newaxis]     # column (alt)

# join
np.concatenate([x, x])           # 1D join
np.vstack([x, x])                # vertical stack
np.hstack([x.reshape(3,1), x.reshape(3,1)])  # horizontal stack

# split
np.split(x, [1, 2])               # split at indices

# 2D split
grid = np.arange(16).reshape(4, 4)
np.vsplit(grid, [2])              # top / bottom
np.hsplit(grid, [2])              # left / right
```
